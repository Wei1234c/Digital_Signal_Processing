Dirac Delta function is : $$\delta(x) = \lim_{a\to 0} \frac{1}{|a|\sqrt{\pi}}e^{-(\frac{x}{a})^2}$$ 

```python
x = np.linspace(-2, 2, 1001)
y = Fourier.dirac_delta_function(x, a = 1e-1)
plt.plot(x, y);
```

```python
N = 64
n = np.arange(N)
ks = np.arange(N)

for k in ks:
#     k = 0
    y = np.exp(-1j * PI2 / N * n * k )  
    # 2Pi 切成 64等分 (n = 64)，每等分 e^(2PI/N*n) 的總和是零。
    # 除非 k = 0，讓 e^(2PI/N*n*k) = e^0 = 1
    
    plt.plot(n, np.cumsum(y))
    
plt.show()
```
